```python
import pandas as pd
from matplotlib import pyplot as plt
import datetime
import json
import os
```

```python

data = os.listdir('./MKL')
data.sort()
```

```python
data
```

```python
new_list = []
for json_file in data[:6]:
    json_file2 = 'MKL/'+json_file
    csv_name = 'mtk-csv/'+json_file.split('.')[0]+'.csv'
    
    with open(json_file2, 'r') as f:
        data2 = json.load(f)
        for i in data2:
            try:
                i['refunds'] = i['refunds'][0]
            except:
                pass
        data3 = pd.json_normalize(data2, max_level = 5)
        a = data3['full_address.shipping_detail.country_code']
        data3 = data3.drop('full_address.shipping_detail.country_code',axis=1)
        data3.insert(0,'full_address.shipping_detail.country_code',a)
        a = data3['order_payment.general_payment_details.payment_total.amount']
        data3 = data3.drop('order_payment.general_payment_details.payment_total.amount',axis=1)
        data3.insert(0,'order_payment.general_payment_details.payment_total.amount',a)

        a = data3['refunds.refund_amount.amount']
        data3 = data3.drop('refunds.refund_amount.amount',axis=1)
        data3.insert(0,'refunds.refund_amount.amount',a)
        a = data3['refunds.merchant_responsible_amount.amount']
        data3 = data3.drop('refunds.merchant_responsible_amount.amount',axis=1)
        data3.insert(0,'refunds.merchant_responsible_amount.amount',a)
        new_list.append(data3)
        
pd.concat(new_list).to_csv('mtk-csv/5aa7861eb722cf6a46c63a32.csv', index=False, encoding='utf_8_sig')       
        
#         data3.to_csv(csv_name, index=False, encoding='utf_8_sig')
```
